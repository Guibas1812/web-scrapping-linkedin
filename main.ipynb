```python
#Block with all job offers info

#Get to Jobs Page Spend in and Retrieve Information
try:
    cookies_accept = driver.find_element(By.XPATH, r'//*[@id="artdeco-global-alert-container"]/div/section/div/div[2]/button[1]')
    cookies_accept.click()
except:
    pass

time.sleep(4)

active_url = driver.current_url

source = requests.get(active_url)
soup = bs4.BeautifulSoup(source.text, 'html.parser')
Page_Counter =  soup.find_all('button')

f = 0 #While loop condition to try
z = 1 #Page Counter

# Loop for getting the information 
while f < 10:
    if z < 42:
        time.sleep(1)
        active_url = driver.current_url
        list = driver.find_elements(By.CSS_SELECTOR,".disabled.ember-view.job-card-container__link.job-card-list__title")
        
        #Press Pagedown button to load all the jobs 
        p = 0
        while p < 20:
            list[0].send_keys(Keys.PAGE_DOWN) 
            p = p + 1  
        
                
        #ApplicationThroughLinkedIn
        card_lists1 = driver.find_elements(By.CSS_SELECTOR, '.job-card-list__footer-wrapper.job-card-container__footer-wrapper.flex-shrink-zero.display-flex.t-sans.t-12.t-black--light.t-normal.t-roman')
        
        for i in range(len(card_lists1)):
            try:
                if 'Easy Apply' in card_lists1[i].text: #Verificiation if easy application through linked in is possible
                    app_linkedin.append(True)
                else:
                    app_linkedin.append(False) #Otherwise appends False 
            except:
                app_linkedin.append(False) #Controll step (in case there is no text)
        
        
        #Promoted
        card_lists = driver.find_elements(By.CSS_SELECTOR, '.job-card-list__footer-wrapper.job-card-container__footer-wrapper.flex-shrink-zero.display-flex.t-sans.t-12.t-black--light.t-normal.t-roman')
        for i in range(len(card_lists)):
            try:
                if 'Promoted' in card_lists[i].text: #Verificiation if job is promoted
                    promoted.append(True)
                else:
                    promoted.append(False) #Otherwise appends false
            except:
                promoted.append(False) #Controll step (in case there is no text)
        
        #Below is a list creation with all info of job's urls classes
        #list = driver.find_elements(By.CSS_SELECTOR,".disabled.ember-view.job-card-container__link.job-card-list__title")
        list_url_1 = driver.find_elements(By.CSS_SELECTOR,".disabled.ember-view.job-card-container__link.job-card-list__title")
        
        urls = []
        url_jobsearch = active_url
        
        #Appends in urls jobs links
        for i in range(len(list_url_1)):    
             urls.append(list_url_1[i].get_attribute('href'))
                
        #Appends every link to urls_2 list - list that will be used for the data frame creation
        #Can't use urls list for the data frame because we need to empty it for code optimization 
        for i in urls:
            urls_2.append(i)
        
        #Driver goes to every link
        for link in urls:        
            driver.get(link)
            time.sleep(2)
            job = requests.get(link)
            job = bs4.BeautifulSoup(job.text, 'html.parser')
            
            #Status
            count = len(status) #Measure created for control procedures
            card_lists2 = driver.find_elements(By.CSS_SELECTOR, '.mt5.mb2') 
            for i in range(len(card_lists2)):
                try:
                    if 'Be an early applicant' in card_lists2[i].text: #Status verificiation 
                        status.append('Early Applicant')
                    elif 'Actively recruiting' in card_lists2[i].text:
                        status.append('On-going')
                    elif 'Actively hiring' in card_lists2[i].text:
                        status.append('On-going')
                    else:
                        status.append('Others')        
                except:
                    status.append(None) #Controll step (in case there is no text)
                    
            count_1 = len(status) #Measure created for control procedures
            if count == count_1: #Control procedure - appends none in case status class was not identified
                status.append(None)
            
            #For some features beautiful soup can be used since it is not necessary to be logged in to gather that info
            
            #JobTitles
            if job.find('h1', {'class':'top-card-layout__title font-sans text-lg papabear:text-xl font-bold leading-open text-color-text mb-0 topcard__title'}) == None:
                job_titles.append(None) #If there is no title
            else:
                job_titles.append(job.find('h1', {'class':'top-card-layout__title font-sans text-lg papabear:text-xl font-bold leading-open text-color-text mb-0 topcard__title'}).text.strip())

            #Companies
            if job.find('span', {'class':'topcard__flavor'}) == None:
                companies.append(None) #If there is no company specification
            else:
                companies.append(job.find('span', {'class':'topcard__flavor'}).text.strip())

            #Location
            if job.find('span', {'class':'topcard__flavor topcard__flavor--bullet'}) == None:
                location.append(None) #If location is not specified
            else:
                location.append(job.find('span', {'class':'topcard__flavor topcard__flavor--bullet'}).text.strip())

            #Dates
            if job.find('span', {'class':'posted-time-ago__text topcard__flavor--metadata'}) == None:
                #As there are two possible dates' classes a seconf if is necessary
                if job.find('span',{'class':'posted-time-ago__text posted-time-ago__text--new topcard__flavor--metadata'}) == None:
                    dates.append(None) #If Date is not specified
                else:
                    dates.append(job.find('span', {'class':'posted-time-ago__text posted-time-ago__text--new topcard__flavor--metadata'}).text.strip())
            else:
                dates.append(job.find('span', {'class':'posted-time-ago__text topcard__flavor--metadata'}).text.strip())
    
            #Applications
            if job.find('span', {'class':'num-applicants__caption topcard__flavor--metadata topcard__flavor--bullet'}) == None:
                #As there are two possible applications' classes a seconf if is necessary
                if job.find('figcaption', {'class':'num-applicants__caption'}) == None:
                    application.append(None) #If application number is not specified
                else: 
                    application.append(job.find('figcaption', {'class':'num-applicants__caption'}).text.strip())
            else:
                application.append(job.find('span', {'class':'num-applicants__caption topcard__flavor--metadata topcard__flavor--bullet'}).text.strip())
            ###In the end of the code, applications' values are gathered only as integers###

            #Python Required
            try:
                if job.find('div', {'class':'show-more-less-html__markup'}).text == None: 
                    text = job.find('div', {'class':'show-more-less-html__markup show-more-less-html__markup--clamp-after-5'}).text #Other possible class
                else:
                    text = job.find('div', {'class':'show-more-less-html__markup'}).text
                
                text = text.upper() #Turns all text in upper letters
                text = text.split() #Creates a list with every word
                if word in text: #if word (that is "PYTHON") is one of those words append True   
                    python_req.append(True)
                else:
                    python_req.append(False)        
            
            except:
                python_req.append(False) #Control procedure in case there is no text
                
            #Employees:
            employees_help = [] #Auxiliar variable creation
            employees1 = [] #Auxiliar variable creation
            time.sleep(1)
            try:
                text_help = driver.find_element(By.XPATH, '/html/body/div[6]/div[3]/div/div[1]/div[1]/div/div[1]/div/div/div[1]/div[2]/ul/li[2]/span').text
                try:
                    help_list = text_help.split('路') #List creation with areas (employees and industry) separated (in case '路' is used)
                    employees_help.append(help_list[0].strip()) #Appends employees number
                except:
                    employees_help.append(None) #Control procedure
            except:
                employees_help.append(None) #Control procedure
            
            try:
                if '-' in employees_help[0]: #in case employees number is specified as "11-50 employees" for example
                    employees1.append(employees_help[0].split('-')[1]) #appends "50 employees" 
                else:
                    employees1.append(employees_help[0]) #When there is no "-"
            except:
                employees1.append(None) #Control procedure
                
            #Only number in employees - string to int  
            for test_string in employees1:
                if test_string == None: #in case number of employees was not specified
                    employees.append(None)      
                else:
                    try:
                        test_string = test_string.replace(',','') #removes "," from number of employees - 10,000 to 10000 for example 
                        test_string = test_string.replace('+','') #for cases when a plus is after the number - indicating that there are more that the specified number
                        helpful_list = [int(i) for i in test_string.split() if i.isdigit()]
                        employees.append(helpful_list[0]) #append only the number
                    except:
                        employees.append(None)
            
            #Workspace
            try:
                workspace.append(driver.find_element(By.CLASS_NAME, "jobs-unified-top-card__workplace-type").text.strip()) #Selenium use to gather workspace
            except:
                workspace.append(None) #When not specified
                
            #Employ_Type and Seniority:
            try:
                #Selenium use to get employment type and security (via XPATH)
                text_help = driver.find_element(By.XPATH, '/html/body/div[6]/div[3]/div/div[1]/div[1]/div/div[1]/div/div/div[1]/div[2]/ul/li[1]/span').text  
                try:
                    help_list = text_help.split('路') #Separates both features
                    employ_type.append(help_list[0].strip()) #appends each feature to correspondent list
                    seniority.append(help_list[1].strip())

                except:
                    seniority.append(None) #when there is an error is because seniority was not indicated
                
            except: #when they are not specified
                employ_type.append(None) 
                seniority.append(None)

            #Industry:
            try:
                #Selenium use to get industry (via XPATH)
                text_help = driver.find_element(By.XPATH, '/html/body/div[6]/div[3]/div/div[1]/div[1]/div/div[1]/div/div/div[1]/div[2]/ul/li[2]/span').text
                try:
                    help_list = text_help.split('路') #splits from employees number
                    industry.append(help_list[1].strip())
                except: #error occurs when industry not specified
                    industry.append(None) 
            except:
                industry.append(None)

            #Bonus
            time.sleep(1)
            followers_bonus_help = [] #auxiliar list creation
            try:
                help_list_bonus = driver.find_element(By.XPATH,'//*[@id="ember25"]') #finds element that contains url
                bonus_url = help_list_bonus.get_attribute('href') #get company's url
                driver.get(bonus_url) #go to that url
                time.sleep(1)
                followers_bonus_help.append(driver.find_element(By.CSS_SELECTOR,".inline-block").text) #appends followers to auxiliar list in string format
                #Only Numbers:
                for test_string in followers_bonus_help: #String to Int
                    if test_string == None:
                        followers_bonus.append(None) #When number of followers is not specified   
                    else:
                        test_string = test_string.replace(',','') #removes "," when needed ("1,000 followers" -> "1000 followers" for example)
                        helpful_list = [int(i) for i in test_string.split() if i.isdigit()] 
                        followers_bonus.append(helpful_list[0]) #appends int
            except: 
                followers_bonus.append(None) #When number of followers is not specified
                             
        driver.get(url_jobsearch)
        time.sleep(1)
        #Switch Pages
        z = z+1
        try:
            time.sleep(1)
            page_X_path = '//*[contains(@aria-label,'+str(z)+')]'
            next_page_button = driver.find_element(By.XPATH, page_X_path)
            next_page_button.click()
        except:
            z = 44
    else:
        f = 11
        
#Only number in applications:
for test_string in application:
    if test_string == None:
        applications.append(None)    
    else:
        helpful_list = [int(i) for i in test_string.split() if i.isdigit()] #auxiliar variable 
        applications.append(helpful_list[0]) #appends int
        
#In "application" they were strings and in "applications" they are int
```

```python
#Create data Frame with the Scrapped information and Write is as CSV file
dict_for_df = {'Job Title':job_titles , 'Company Name':companies, 'Location':location,'Posting Date':dates,
               'Number of Applicants':applications, 'Offer URL':urls_2, 'Seniority':seniority, 
               'Employment Type':employ_type, 'Industry':industry, 'Python Required':python_req,
               'Status':status, 'Workspace':workspace, 'Number of Employees':employees, 
               'Application through Linkedin':app_linkedin, 'Promoted':promoted, 'Followers':followers_bonus}
    

df = pd.DataFrame(dict_for_df)

df.head(50) #only showing first 50 results 
```
